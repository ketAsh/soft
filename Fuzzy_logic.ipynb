# Question A


```python
def membership_A(x):
    if x < 30:
        return 1
    elif 30 <= x <= 50:
        return (x - 30) / (50 - 30)
    elif 50 < x <= 70:
        return (x - 50) / (70 - 50)
    else:
        return 0
```

```python
def membership_B(x):
    if x < 60:
        return 0
    elif 60 <= x <= 80:
        return (x - 60) / (80 - 60)
    elif 80 < x <= 100:
        return (x - 80) / (100 - 80)
    else:
        return 1
```

```python
membership_A(40)
```

```python
speed_values = [1,30,33,66,77,88,97,109,200]

membership_values_A = []
membership_values_B = []

for x in speed_values:
    a = membership_A(x);
    b = membership_B(x);
    
    membership_values_A.append(a)
    membership_values_B.append(b)
```

```python
membership_values_A, membership_values_B
```

```python
membership_max_C = [max(membership_values_A[x],membership_values_B[x]) for x in range(len(speed_values))]
membership_min_D = [min(membership_values_A[x],membership_values_B[x]) for x in range(len(speed_values))]
```

```python
membership_max_C, membership_min_D
```

```python
plt.figure(figsize=(10,6))
plt.plot(speed_values,membership_values_A)
```

```python
plt.plot(speed_values,membership_values_B)
```

```python
plt.plot(speed_values,membership_max_C)
```

```python
plt.plot(speed_values,membership_min_D)
```

# Question B


```python
class temperatureMemebershipFunction:
    def C(x):
        if x < -10:
            return 0
        elif x > -10 and x < 0:
            return (x - (-10))/(0-(-10))
        elif x > 0 and x < 5:
            return 1
        else:
            return 0
    
    def D(x):
        if x < 25:
            return 0
        elif x < 35:
            return (x - 25)/(35 - 25)
        elif x < 40:
            return 1
        else:
            return 0
```

```python
# universeOfDiscourseTemperature = []
# for i in range(-20, 50):
    
#     universeOfDiscourseTemperature.append(i + 2)
# universeOfDiscourseTemperature
universeOfDiscourseTemperature = []
for i in range(-2, 11):
    universeOfDiscourseTemperature.append(i * 5)
universeOfDiscourseTemperature
```

```python
membership_values_C = []
membership_values_D = []

for x in universeOfDiscourseTemperature:
    g = temperatureMemebershipFunction
    c = g.C(x)
    d = g.D(x)
    membership_values_C.append(c)
    membership_values_D.append(d)
```

```python
membership_values_C, membership_values_D
```

```python
matrix_C = np.matrix(membership_values_C)
matrix_D = np.matrix(membership_values_D).transpose()
```

```python
matrix_C, matrix_D
```

```python
size = len(membership_values_C)
m = np.zeros((size,size))
max_min = []
for i in range(size):
    for j in range(size):
        m[i][j] = min(matrix_C[0,j], matrix_D[i,0]);
    max_min.append(max(m[i]))
    
max_min
```

```python
max_min
```

```python
plt.plot(universeOfDiscourseTemperature,max_min)
plt.title("membership curve of F(x) = max (min(C(x), D(x)))")
```
